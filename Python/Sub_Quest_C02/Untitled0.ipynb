```python
# Avengers 파일 불러오기

with open('/content/drive/MyDrive/Colab/Avengers.txt', 'r') as file:
  content = file.read()

print(content)
```

```python
lowwer_content = content.lower() ##불러온 문서 내 대문자를 소문자로 바꾸기
print(lowwer_content)
```

```python
import string ##챗GPT를 사용한 특수문자 지우기 시도
'lowwer_content'.strip(string.punctuation)
'lowwer_content'
string.punctuation
'!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'

strip_content = ''.join(char for char in lowwer_content if char not in string.punctuation)

print(strip_content)

## 좋은 공부가 되었기 때문에 흔적만 남기고 다시 시도
```

```python
list_content = lowwer_content.split()
print(list_content)
```

```python
import string ## 배운 내용을 바탕으로 한 특수문자 지우기

Strip_content = [item.strip(string.punctuation) for item in list_content]

print(Strip_content)
```

```python
# 단어 빈도수 계산
count_dic = {word: Strip_content.count(word) for word in set(Strip_content)}

print(count_dic)
```

```python
# 딕셔너리에서 공백 삭제

del count_dic['']

print(count_dic)
```

```python
# 딕셔너리를 빈도수 순으로 내림차순 정열

dic = dict(sorted(dic.items(), key=lambda item: item[1], reverse=True))

print(dic)
```

```python
# 내림차순 된 순위대로 정수 인덱스 부여 (rank 함수 사용)

ranked_dic = {rank: {'word': word, 'count': count} for rank, (word, count) in enumerate(dic, start=1)}

print(ranked_dic)

# 오류 발생
```

```python
# 내림차순 된 순위대로 정수 인덱스 부여 (index 함수 사용)

vocab = {index: key for key, index in enumerate(dic.keys(), start=0)}

print(vocab)
```

```python
# key 값을 input 했을 때, key 값에 맞는 정수를 return

text = input('찾으시고자 하는 단어를 입력하세요.')

words = text.split()

encoded_values = []
for word in words:
  if word in vocab:
        encoded_values.append(vocab[word])
  else:
        print('해당 단어를 찾을 수 없습니다.')

print(encoded_values)
```
