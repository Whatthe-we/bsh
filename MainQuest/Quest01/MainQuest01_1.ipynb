```python
#메인퀘스트 1번 : 은행계좌 만들기(3점)
#version.1.0.8
"""
은행 계좌개설 기능 및 계좌개설로직 구현 -> 실행시 메뉴 출력으로 변경, 계좌개설 메뉴 이외 기능 미구현
-> 입금기능 계좌검색 기능 및 이자 구현 -> 출금기능 구현 -> 관리자 메뉴 접근 기능 구현 -> 관리자 메뉴 마무리
"""
import random

class Account:
    # 클래스 변수로 은행 이름을 고정
    bank_name = "SC은행"

    # 클래스 변수로 계좌번호 생성을 위한 카운터 초기화
    account_counter = 1

    def __init__(self, owner_name, initial_balance):

        # 예금주 이름 설정
        self.owner_name = owner_name

        # 초기 잔액 설정 (음수 방지를 위한 최소 0 처리)
        self.balance = max(0, initial_balance)

        # 계좌번호 자동 생성
        # 순서대로 증가하는 카운터와 랜덤 요소를 결합
        self.account_number = self._generate_account_number()

        # 입금 횟수 초기화
        self.deposit_count = 0

        # 입출금 기록 저장
        self.transaction_history = []

    def _generate_account_number(self):

        # 첫 번째 세그먼트: 1~999 사이의 랜덤 숫자
        first_segment = random.randint(1, 999)

        # 두 번째 세그먼트: 10~99 사이의 랜덤 숫자
        second_segment = random.randint(10, 99)

        # 세 번째 세그먼트: 1~999999 사이의 랜덤 숫자
        third_segment = random.randint(1, 999999)

        # 계좌번호 형식: 123-12-123456
        return f"{first_segment:03d}-{second_segment:02d}-{third_segment:06d}"

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.deposit_count += 1
            self.transaction_history.append(f"입금: +{amount:,}원 (잔액: {self.balance:,}원)")
            print(f"{amount:,}원이 입금되었습니다. 현재 잔액: {self.balance:,}원")

            # 5번째 입금마다 이자 지급
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                self.transaction_history.append(f"이자 지급: +{interest:.0f}원 (잔액: {self.balance:,}원)")
                print(f"이자가 지급되었습니다! 지급된 이자: {interest:.0f}원, 현재 잔액: {self.balance:,}원")
        else:
            print("1원 이상의 금액을 입금해주세요.")

    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                self.transaction_history.append(f"출금: -{amount:,}원 (잔액: {self.balance:,}원)")
                print(f"{amount:,}원이 출금되었습니다. 현재 잔액: {self.balance:,}원")
            else:
                print("잔액이 부족합니다.")
        else:
            print("1원 이상의 금액을 입력해주세요.")

    def __str__(self):
        return f"은행: {self.bank_name}, 예금주: {self.owner_name}, 계좌번호: {self.account_number}, 잔액: {self.balance:,}원"

# 사용자로부터 계좌 정보를 입력받아 Account 객체를 생성하는 함수
def create_account():
    print("=== SC은행 계좌 개설 ===")

    # 예금주 이름 입력
    while True:
        owner_name = input("예금주 이름을 입력하세요: ").strip()
        if owner_name:
            break
        print("이름을 입력해주세요.")

    # 초기 잔액 입력
    while True:
        try:
            initial_balance = int(input("초기 입금액을 입력하세요: "))
            if initial_balance >= 0:
                break
            print("0 이상의 금액을 입력해주세요.")
        except ValueError:
            print("숫자로 입력해주세요.")

    # Account 객체 생성 및 반환
    return Account(owner_name, initial_balance)

def deposit_to_account(accounts):
    print("\n입금을 선택하셨습니다.")
    while True:
        account_number = input("입금할 계좌번호를 입력하세요: ").strip()
        # 입력받은 계좌번호로 계좌 검색
        account = next((acc for acc in accounts if acc.account_number == account_number), None)
        if account:
            print(f"예금주: {account.owner_name}")
            confirm = input("이 계좌에 입금하는 것이 맞습니까? (Y/N): ").strip().lower()
            if confirm == 'y':
                break
            else:
                print("계좌 선택을 취소했습니다. 다시 입력해주세요.")
        else:
            print("일치하는 계좌번호가 없습니다. 다시 시도해주세요.")

    while True:
        try:
            amount = int(input("입금 금액을 입력해주세요 (최소 1원): "))
            if amount >= 1:
                account.deposit(amount)
                print(f"입금이 완료되었습니다. 예금주: {account.owner_name}, 현재 잔액: {account.balance:,}원")
                break
            else:
                print("1원 이상의 금액을 입력해주세요.")
        except ValueError:
            print("숫자로 입력해주세요.")

def withdraw_from_account(accounts):
    print("\n출금을 선택하셨습니다.")
    while True:
        account_number = input("출금할 계좌번호를 입력하세요: ").strip()
        # 입력받은 계좌번호로 계좌 검색
        account = next((acc for acc in accounts if acc.account_number == account_number), None)
        if account:
            while True:
                owner_name = input("예금주 이름을 입력하세요: ").strip()
                if owner_name == account.owner_name:
                    print("계좌와 예금주가 확인되었습니다.")
                    break
                else:
                    print("다시 확인해주세요.")
            break
        else:
            print("일치하는 계좌번호가 없습니다. 다시 시도해주세요.")

    while True:
        try:
            amount = int(input("출금 금액을 입력해주세요 (최소 1원): "))
            if amount >= 1:
                account.withdraw(amount)
                break
            else:
                print("1원 이상의 금액을 입력해주세요.")
        except ValueError:
            print("숫자로 입력해주세요.")

def admin_menu(accounts):
    print("\n관리자 접근이 필요합니다. 초기 비밀번호는 0000입니다.")
    while True:
        password = input("관리자 비밀번호를 입력하세요: ").strip()
        if password == "0000":
            print("\n관리자 메뉴에 접근합니다.")
            while True:
                print("\n관리자 메뉴:")
                print("1. 계좌정보출력")
                print("2. 입출금 기록 열람")
                print("3. VIP정보")
                print("4. 종료")
                choice = input("선택: ").strip()
                if choice == "1":
                    print("\n=== 계좌 정보 출력 ===")
                    print(f"총 계좌 개수: {len(accounts)}개")
                    for account in accounts:
                        print(account)
                elif choice == "2":
                    print("\n=== 입출금 기록 열람 ===")
                    for account in accounts:
                        print(f"계좌번호: {account.account_number}, 예금주: {account.owner_name}")
                        for record in account.transaction_history:
                            print(f"  {record}")
                elif choice == "3":
                    print("\n=== VIP 정보 ===")
                    vip_accounts = [account for account in accounts if account.balance >= 1000000]
                    print(f"VIP 계좌 총 개수: {len(vip_accounts)}개")
                    for account in vip_accounts:
                        print(account)
                elif choice == "4":
                    print("관리자 메뉴를 종료합니다.")
                    break
                else:
                    print("잘못된 선택입니다. 다시 시도해주세요.")
            break
        else:
            print("비밀번호가 틀렸습니다. 다시 시도해주세요.")

def main():
    print("어서오세요 SC은행입니다. 무엇을 도와드릴까요?")

    # 계좌 목록 관리
    accounts = []

    while True:
        # 메뉴 출력
        print("\n메뉴를 선택해주세요:")
        print("1. 계좌개설")
        print("2. 입금")
        print("3. 출금")
        print("4. 관리자")
        print("5. 종료")

        # 사용자 입력 받기
        choice = input("메뉴 선택: ").strip()

        if choice == "1":
            # 계좌 개설 기능 실행
            new_account = create_account()
            accounts.append(new_account)
            print("\n### 계좌 개설 완료 ###")
            print(new_account)
        elif choice == "2":
            # 입금 기능 실행
            if accounts:
                deposit_to_account(accounts)
            else:
                print("등록된 계좌가 없습니다. 먼저 계좌를 개설해주세요.")
        elif choice == "3":
            # 출금 기능 실행
            if accounts:
                withdraw_from_account(accounts)
            else:
                print("등록된 계좌가 없습니다. 먼저 계좌를 개설해주세요.")
        elif choice == "4":
            # 관리자 메뉴 실행
            if accounts:
                admin_menu(accounts)
            else:
                print("등록된 계좌가 없습니다. 먼저 계좌를 개설해주세요.")
        elif choice == "5":
            # 프로그램 종료
            print("SC은행을 이용해 주셔서 감사합니다. 안녕히 가세요!")
            break
        else:
            print("준비 중인 기능이거나 잘못된 선택입니다. 다시 시도해주세요.")

# 프로그램 실행
if __name__ == "__main__":
    main()
```
